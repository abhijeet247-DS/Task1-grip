## Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sb
```

## Reading Dataset

```python
df=pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
df.head()
```

## Visualizing Dataset

```python
sb.swarmplot(x="Hours",y="Scores",data=df,palette="Set1",split=True)
```

```python
sb.lmplot(x="Hours",y="Scores",height=3,aspect=6,data=df)
```

## Dividing the data between labels and attributes

```python
x=df.iloc[:, :-1].values
y=df.iloc[:, 1].values
```

## Training of model

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```

## Algorithm-Linear Regression

I have choosen simple Linear Regression algorithm as the dataset only contains two columns:
Hours and Scores(which implies that the dataset has one dependent and one independent variable) and its linear(as per the visualizations)

```python
from sklearn.linear_model import LinearRegression
a=LinearRegression()
a.fit(xtrain,ytrain)
```

## Testing algorithm

```python
ypred=a.predict(xtest)
df1= pd.DataFrame({'Actual': ytest, 'Predicted': ypred})  
df1
```

## Checking the accuracy of algorithm

```python
from sklearn.metrics import r2_score
r2_score(ytest,ypred)
```

## Predicting score for 9.25 hours

```python
Hours =[[9.25]] 
own_pred = a.predict(Hours)
print("No of Hours = {}".format(Hours))
print("Predicted Score = {}".format(own_pred[0]))
```
